```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the  dataset csv file 
df = pd.read_csv('country_wise_latest.csv')

df.head() # printing first 5 rows 

```

```python
#checking and cleaning dataset 
print(df.isnull().sum())

# Using forward fill \
df.ffill(inplace=True)  # Forward fill missing values
# Verify missing values are handled
print(df.isnull().sum())

```

```python
#doing statiscal analysis and applying matplotlib and seaborn  for Data visualization
top_countries = df.nlargest(10, 'Confirmed')


# Plot pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_countries['Confirmed'], labels=top_countries['Country/Region'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries - Percentage of COVID Cases')
plt.show()



region_cases = df.groupby('WHO Region')['New cases'].sum().reset_index()
# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='WHO Region', y='New cases', hue='WHO Region', data=region_cases, palette='coolwarm', legend=False)
plt.xlabel('WHO Region')
plt.ylabel('Total New Cases')
plt.title('New COVID Cases by WHO Region')
plt.xticks(rotation=45)
plt.show()

#plotting  histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Recovered'], bins=20, color='green', edgecolor='black')
plt.xlabel('Recovered Cases')
plt.ylabel('Number of Countries')
plt.title('Distribution of Recovered Cases')
plt.show()


```
